```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
#setosa
SSlength = []
SSwidth = [] 
SPlength= [] 
SPwidth= []
```

```python
#versicolor
siSlength = []
siSwidth = [] 
siPlength= [] 
siPwidth= []
```

```python
#verginica
giSlength = []
giSwidth = [] 
giPlength= [] 
giPwidth= []
```

```python
i=0
f = open('iris.csv', 'r', encoding='utf-8')
iris = csv.reader(f)
for line in iris: #data classification
    if i<50 : #setosa
        SSlength.append(float(line[0])) #sepal length
        SSwidth.append(float(line[1])) #sepal width
    
        SPlength.append(float(line[2])) #petal length
        SPwidth.append(float(line[3]))  #petal width
    
    elif i<100 : #versicolor
        siSlength.append(float(line[0])) #sepal length
        siSwidth.append(float(line[1])) #sepal width
    
        siPlength.append(float(line[2])) #petal length
        siPwidth.append(float(line[3]))  #petal width
    
    else : #verginica
        giSlength.append(float(line[0])) #sepal length
        giSwidth.append(float(line[1])) #sepal width
    
        giPlength.append(float(line[2])) #petal length
        giPwidth.append(float(line[3]))  #petal width
    i+=1
f.close()
```

```python
Sscatter = plt.scatter(SSlength,SSwidth) #bout sepal setosa
siscatter = plt.scatter(siSlength, siSwidth) #bout sepal versicolor
giscatter = plt.scatter(giSlength, giSwidth) #bout sepal virginica

plt.xlabel('Sepal length', color='white')
plt.ylabel('Sepal width', color='white')
```

```python
Sscatter = plt.scatter(SPlength,SPwidth) #bout petal setosa
siscatter = plt.scatter(siPlength, siPwidth) #bout petal versicolor
giscatter = plt.scatter(giPlength, giPwidth) #bout petal virginica

plt.xlabel('Petal length', color='white')
plt.ylabel('Petal width', color='white')
```
